```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
data1 = pd.read_csv('/content/ad-clicks.csv')
data2 = pd.read_csv('/content/buy-clicks.csv')
data3 = pd.read_csv('/content/game-clicks.csv')
data4 = pd.read_csv('/content/user-session.csv')
combined = pd.read_csv('/content/combined-data.csv')
```

```python
#mrging my datasets to create my own
merged_df = pd.merge(data1, data3, on='userSessionId',sort=True,suffixes=('_x', '_y'))
merged = pd.merge(merged_df, data2, on='userSessionId',sort=True,suffixes=('_x', '_y'))
merged_final = pd.merge(merged, data4, on='userSessionId',sort=True,suffixes=('_x', '_y'))
merged_final = merged_final.drop(['timestamp_x','timestamp_y','txId_x','teamId_x','userId_x','teamLevel_x','userId_y','teamId_y','txId_y','teamLevel_y','buyId'], axis=1)
merged_final.drop_duplicates(subset=['clickId'])
# Print the merged DataFrame
merged_final
```

```python
from sklearn.preprocessing import LabelEncoder
temp = merged_final
enc = LabelEncoder()
label_encoder = enc.fit(temp['platformType'])
print ("Categorical classes:", label_encoder.classes_)
integer_classes = label_encoder.transform(label_encoder.classes_)
print ("Integer classes:", integer_classes)
t = label_encoder.transform(temp['platformType'])
temp['EncPlatformType'] = t
```

```python
enc = LabelEncoder()
label_encoder = enc.fit(temp['sessionType'])
print ("Categorical classes:", label_encoder.classes_)
integer_classes = label_encoder.transform(label_encoder.classes_)
print ("Integer classes:", integer_classes)
t = label_encoder.transform(temp['sessionType'])
temp['EncsessionType'] = t
```

```python
enc = LabelEncoder()
label_encoder = enc.fit(temp['adCategory'])
print ("Categorical classes:", label_encoder.classes_)
integer_classes = label_encoder.transform(label_encoder.classes_)
print ("Integer classes:", integer_classes)
t = label_encoder.transform(temp['adCategory'])
temp['EncadCategory'] = t
```

```python
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn import neighbors

clf = tree.DecisionTreeClassifier()

clf1 = svm.SVC()
clf2 = GaussianNB()
clf3 = neighbors.KNeighborsClassifier()
#adCategory	clickId	isHit	team	price	teamId	assignmentId	sessionType	platformType
X = temp[[ 'price','EncPlatformType' , 'EncsessionType']][0:20000]

Y = temp['EncadCategory'][0:20000]


#CHALLENGE - ...and train them on our data
clf = clf.fit(X , Y)

#Support Vector classifier
clf1 = clf1.fit(X,Y)

#Naive Bayes
clf2 = clf2.fit(X,Y)

#K neighbors classifier
clf3 = clf3.fit(X,Y)

#test data
X_test = temp[[ 'price','EncPlatformType' , 'EncsessionType']][20000:30000]

Y_test = temp['EncadCategory'][20000:30000]


Y_prediction  = clf.predict(X_test)
Y_prediction1 = clf1.predict(X_test)
Y_prediction2 = clf2.predict(X_test)
Y_prediction3 = clf3.predict(X_test)

#CHALLENGE compare their reusults and print the best one!

print("Prediction for Naive Bayes : ",Y_prediction2)

print("Accuracy for Decision Tree : ",accuracy_score(Y_test,Y_prediction))

print("Accuracy for SVM : ",accuracy_score(Y_test,Y_prediction1))

print("Accuracy for K neighbors : ",accuracy_score(Y_test,Y_prediction3))
```

```python
import numpy as np
Y_prediction  = clf.predict(X_test)
np.unique(Y_prediction)
pinkFlam_evaluation = temp[20000:30000]
pinkFlam_evaluation['Predicted_adCategory'] = Y_prediction
pinkFlam_evaluation[['Predicted_adCategory' , 'EncadCategory' ]] 
y_true = pinkFlam_evaluation['EncadCategory'] 
y_pred = pinkFlam_evaluation['Predicted_adCategory']
```

```python
from sklearn.metrics import confusion_matrix
tn, fp, fn, tp = confusion_matrix([0, 1, 0, 1], [1, 1, 1, 0]).ravel()
```

```python
unique_label = np.unique(y_true)
print(pd.DataFrame(confusion_matrix(y_true, y_pred, labels=unique_label), 
                   index=['true:{:}'.format(x) for x in unique_label], 
                   columns=['pred:{:}'.format(x) for x in unique_label]))
```

```python
cm = pd.DataFrame(confusion_matrix(y_true, y_pred, labels=unique_label), 
                   index=['true:{:}'.format(x) for x in unique_label], 
                   columns=['pred:{:}'.format(x) for x in unique_label])

sns.heatmap(cm, annot=True,
         fmt='.2f',
         xticklabels = ["Left", "No Left"] , yticklabels = ["Left", "No Left"] )

plt.ylabel('True User Category')
plt.xlabel('Predicted User Category')
```
